## 비중 생성

### 예전 Ver

```python
random.seed(1)
test_e = list()
test_v = list()
want_v = list()
want_e = list()
Iter = 1
port_dict = {}
for j in range(1000000):
    w = list(range(14))
    for i in range(13):
        if i == 0 :
            temp = random.uniform(0,1)*0.4
            w[i] = temp
        else :
            temp = random.uniform(0,1 - sum(w[:i]))*0.4
            w[i] = temp
    w[13] = 0.99 - sum(w[:-1])
    for k in range(1):
        w = list(map(lambda x : round(x,4), w))
        if(w[0] <= 0.4 and w[1] <= 0.2 and 0.1 <= w[0] + w[1] and w[0] + w[1] <= 0.4 and w[2] <= 0.2 and w[3] <= 0.2 and w[4] <= 0.2 and w[5] <= 0.2 and w[2]+w[3]+w[4]+w[5] < 0.4 and w[6] <= 0.5 and w[7] <= 0.4 and 0.05<w[8] and w[8] < 0.4 and 0.2 < w[6] + w[7] + w[8] and w[6] + w[7] + w[8] < 0.6 and w[9] <= 0.15 and w[10] <= 0.15 and 0.05 < w[9] + w[10] and w[9] + w[10] < 0.2 and w[11] <= 0.2 and w[12] <= 0.2 and w[11] + w[12] <= 0.2 and w[12] <= 0.5):
            test_e.append(np.dot(np.mean(newdata_150),np.array(w)))
            test_v.append(np.var(np.dot(newdata_150,w)))
            if 0.047 < test_e[-1] and test_v[-1] < 0.0045 :
                want_v.append(test_v[-1])
                want_e.append(test_e[-1])
                port_dict[Iter] = w
                Iter = Iter + 1
                print('final', w)
```

- 위 결과의 각 portfolio 후보에 대한 추이 비교

```python
plt.figure(figsize = (12,9))
for i in range(1, len(port_dict)+1):
    plt.plot(range(len(port_dict[i])), port_dict[i])
plt.show()
```

```python
plt.figure(figsize = (12,9))
plt.plot(test_v,test_e,'.')
plt.plot(want_v,want_e,'o')
plt.show()
```

### 현재 Ver

and 0.1 <= w[0] + w[1] 조건 제외!

```python
random.seed(1)
# 대회 조건에 맞는 후보 1차 portfolio filtering
test_e_s = list()
test_v_s = list()
# 팀이 원하는 조건1에 맞는 후보 portolio 2차 filtering
want_v_s = list()
want_e_s = list()
# 팀이 원하는 조건2에 맞는 후보 portolio 2차 filtering
want_v_s_2 = list()
want_e_s_2 = list()
Iter = 1
port_dict_s = {}
for j in range(100000):
    w = list(range(14))
    for i in range(13):
        if i == 0 :
            temp = random.uniform(0,1)*0.4
            w[i] = temp
        else :
            temp = random.uniform(0,1 - sum(w[:i]))*0.4
            w[i] = temp
    w[13] = 0.99 - sum(w[:-1])
    for k in range(10):
        w = random.sample(w, len(w))
        w = list(map(lambda x : round(x,4), w))
        if(w[0] <= 0.4 and w[1] <= 0.2  and w[0] + w[1] <= 0.4 and w[2] <= 0.2 and w[3] <= 0.2 and w[4] <= 0.2 and w[5] <= 0.2 and w[2]+w[3]+w[4]+w[5] < 0.4 and w[6] <= 0.5 and w[7] <= 0.4 and 0.05<w[8] and w[8] < 0.4 and 0.2 < w[6] + w[7] + w[8] and w[6] + w[7] + w[8] < 0.6 and w[9] <= 0.15 and w[10] <= 0.15 and 0.05 < w[9] + w[10] and w[9] + w[10] < 0.2 and w[11] <= 0.2 and w[12] <= 0.2 and w[11] + w[12] <= 0.2 and w[12] <= 0.5):
            test_e_s.append(np.dot(np.mean(newdata_150),np.array(w)))
            test_v_s.append(np.var(np.dot(newdata_150,w)))
            if 0.045 < test_e_s[-1] and test_v_s[-1] < 0.003 :
                want_v_s.append(test_v_s[-1])
                want_e_s.append(test_e_s[-1])
                port_dict_s[Iter] = w
                Iter = Iter + 1
                print('shuffled_final1', w)
            elif 0.05 < test_e_s[-1] and test_v_s[-1] < 0.0043 :
                want_v_s_2.append(test_v_s[-1])
                want_e_s_2.append(test_e_s[-1])
                port_dict_s[Iter] = w
                Iter = Iter + 1
                print('shuffled_final2',w)
```

- 위 결과의 각 portfolio 후보에 대한 추이 비교

```python
plt.figure(figsize = (12,9))
for i in range(1, len(port_dict_s)+1):
    plt.plot(list(map(lambda x : x[:7],list(newdata_150.columns))), port_dict_s[i])
plt.show()
```

```python
plt.figure(figsize = (12,9))
plt.plot(test_v_s,test_e_s,'.')
plt.plot(want_v_s,want_e_s,'o')
plt.plot(want_v_s_2,want_e_s_2, 'o')
plt.show()
```

## 옛날 Ver과 현재 Ver Layering한 Plot

```python
plt.figure(figsize = (12,9))
plt.plot(test_v_s,test_e_s,',') # suffle한 1차 portfolio 후보 (Blue)
plt.plot(test_v,test_e,',') # suffle하지 않은 1차 portfolio 후보 (Orange)
plt.plot(want_v,want_e,'o')
plt.plot(want_v_s, want_e_s,'o')
plt.plot(want_v_s_2,want_e_s_2, 'o')
plt.show()
```
